```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(25)
num_of_points = 1000
height = np.random.randint(140,215,num_of_points)
weight = np.random.randint(40,130,num_of_points)
```

```python
df = pd.DataFrame({'height':height,'weight':weight,'BMI':None,'bodytype':None})
```

```python
df.head()
```

```python
df['BMI'] = df['weight']/((df['height']/100)**2)
df['bodytype'] = df['BMI'].map(lambda x: 'thin' if x<18 else 'normal' if x<30 else 'obese')
```

```python
df.head(10)
```

```python
sns.set_context('notebook')
sns.scatterplot(df.height,df.weight,hue=df.bodytype)
```
