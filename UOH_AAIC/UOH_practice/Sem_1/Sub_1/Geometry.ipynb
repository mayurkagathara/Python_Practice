```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
from ipywidgets import *
%matplotlib inline
```

```python
np.random.seed(75)
sample = 50
data_afr = [np.random.normal(180.0,5.0,sample),np.random.normal(80.0,5.0,sample)] #African national Weight, Height
data_chn = [np.random.normal(165.0,5.0,sample),np.random.normal(70.0,5.0,sample)] #Chinese national Weight, Height
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.set(title='2d data visulization',xlabel='Height', ylabel='Weight',xlim=(150, 200), ylim=(50, 100))

ax.scatter(data_afr[0],data_afr[1], c=['green'], s=60, edgecolor='black',label='African')
ax.scatter(data_chn[0],data_chn[1], c=['Red'], s=60, edgecolor='black',label='Chinese')



ax.legend()
plt.show()
```

### Interactive plot to find the value of slope and intersection

```python
x = np.linspace(150,200,50)


def update(m=-0.99,c=246):
    y = c + m*x
    plt.scatter(data_afr[0],data_afr[1], c=['green'], s=[60], edgecolor='black',label='African')
    plt.scatter(data_chn[0],data_chn[1], c=['Red'], s=[60], edgecolor='black',label='Chinese')
    plt.plot(x,y,'b-')
    plt.show()
    
interactive_plot = interactive(update, m=(-2,2,0.01), c=(150,1000,0.1))
interactive_plot
```

### Let's plot the line in the data

```python
plt.scatter(data_afr[0],data_afr[1], c=['green'], s=60, edgecolor='black',label='African')
plt.scatter(data_chn[0],data_chn[1], c=['Red'], s=60, edgecolor='black',label='Chinese')

x = np.linspace(150,200,50)
c = 246
m = -0.99
y = c + m*x

line_equation=f"y =({m})x+{c}"

plt.plot(x,y,'b-',)

plt.title('Line separating categories')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.annotate(line_equation,xy=(190,58), xytext=(180, 50),
#              arrowprops=dict(facecolor='black', shrink=0.05)
             arrowprops=dict(arrowstyle="->")
            )
plt.legend()
plt.show()
```

### Plane: ax + by + cz + d = 0

where normal is (a,b,c)  
if we take (x,y) = (0,0) we can get a point -d/c on a plane.  
using normal and point (0, 0, -d/c) we can calculate d.  
d = -ax -by -cz  
let's code this.  

```python
# plane: -4x + 5y + 9z - 9 = 0
def updateplane(a=-3290.0,b=-7650.0,c=-840.0,d=1168850.0):
    point  = np.array([0,0,(-d/c)])
    normal = np.array([a,b,c])
    d = -np.sum(point*normal)# dot product
    xx, yy = np.meshgrid(range(150,190,1), range(60,90,1))
    z3 = (-normal[0]*xx - normal[1]*yy - d)/normal[2]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data_afr[0], data_afr[1], data_afr[2], marker='o')
    ax.scatter(data_chn[0], data_chn[1], data_chn[2], marker='^')
    ax.plot_surface(xx,yy,z3, color='cyan',alpha=0.5)

    plt.show()

    
interactive_plot = interactive(updateplane, a=(-100000,100000,0.1), b=(-100000,100000,0.1), c=(-100000,100000,0.1), d=(-100000000,100000000,0.1))
interactive_plot
```
